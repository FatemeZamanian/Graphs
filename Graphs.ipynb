```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
N=1000
std=0.6

banana_width=np.random.normal(4,std,N)
banana_height=np.random.normal(8,std,N)

apple_width=np.random.normal(6,std,N)
apple_height=np.random.normal(6,std,N)

width=np.concatenate((banana_width,apple_width))
height=np.concatenate((banana_height,apple_height))

x=np.array([width,height]).T

apple_label=np.zeros(N,dtype='int')
banana_label=np.ones(N,dtype='int')
Y=np.concatenate((apple_label,banana_label))
```

```python
plt.figure()
plt.scatter(x[0:N,0],x[0:N,1], c='blue')
plt.scatter(x[N:,0],x[N:,1], c='orange')

plt.xlabel('widths')
plt.ylabel('heights')
plt.show()
```

```python
labels = ['bananas', 'apples', 'bananas', 'apples']
x = [1, 1.5, 2.5, 3.0]
height = [banana_width.mean(),apple_width.mean(),banana_height.mean(),apple_height.mean()]
yerr = [banana_width.std(),apple_width.std(),banana_height.std(),apple_height.std()]
color = ['tab:blue', 'tab:blue', 'tab:orange', 'tab:orange']

plt.bar(x=x,
        height=height,
        color=color,
        width=0.4,
        error_kw={'lw': 4},
        yerr=yerr)
plt.xticks(ticks=x, labels=labels)
plt.title('Fruit length VS width')

plt.grid()
```

```python
plt.boxplot(x=banana_height, patch_artist=True, boxprops={'facecolor': 'tab:blue'}, positions=[0], labels=['bananas'], widths=[0.2])
plt.boxplot(x=apple_height, patch_artist=True, boxprops={'facecolor': 'tab:pink'}, positions=[0.2], labels=['apple'], widths=[0.2])

plt.ylabel('heights')
plt.xlim(-0.2, 0.5)
plt.title('length Distribution ')
```

```python
plt.hist(x=banana_height,bins=100, density=True, color='tab:blue', label='banana',)
plt.hist(x=apple_height,bins=100, density=True, color='tab:pink', label='apple', alpha=0.8)
plt.legend(frameon=False)
plt.title('height Distribution ')
plt.xlabel('height')
```

```python
fig, (ax, cax) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [0.9, 0.1]})
_, _, _, mappable = ax.hist2d(
    x=height, y=width, bins=30, cmap=plt.cm.hot)
ax.set_xlabel('height')
ax.set_ylabel('width')

plt.colorbar(mappable=mappable, cax=cax, ax=ax)
```

```python
b= banana_height/banana_width
a = apple_height/apple_width

plt.hist(x=b,
         bins=50, density=True, color='tab:blue', label='banana',)
plt.hist(x=a,
         bins=50, density=True, color='tab:pink', label='apple',)
plt.legend()
plt.title('L/W Distribution ')
plt.xlabel('L/W ratio')
```

```python
b=banana_width*banana_height
a=apple_width*apple_height
plt.hist(x=b, histtype='step', lw=4,
         bins=40, density=True, color='tab:blue', label='bananas')
plt.hist(x=a, histtype='step', lw=4,
         bins=40, density=True, color='tab:pink', label='apples')
plt.title('Area Distribution ')
plt.legend()
plt.xlabel('Area')
```

```python
b = 2*(banana_height+banana_width)
a = 2*(apple_height+apple_width)

plt.hist(x=b, histtype='step', lw=4,
         bins=50, density=True, color='tab:blue', label='bananas')
plt.hist(x=a, histtype='step', lw=4,
         bins=50, density=True, color='tab:pink', label='apples')
plt.title('circumference Distribution ')
plt.legend()
plt.xlabel('Circumference')
```
